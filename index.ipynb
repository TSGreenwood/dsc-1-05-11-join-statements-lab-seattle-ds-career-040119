## Connecting to the Database
Import the necessary packages and connect to the database **data.sqlite**.

```python
import sqlite3
conn = sqlite3.connect("data.sqlite", detect_types=sqlite3.PARSE_COLNAMES)
cur = conn.cursor()
import pandas as pd
```

```python
cur.execute("""SELECT * FROM orderdetails 
            JOIN products 
            ON orderdetails.productCode = products.productCode
            LIMIT 20;""")
d_ord_ds_prod = pd.DataFrame(cur.fetchall())
d_ord_ds_prod.columns = [i[0] for i in cur.description]
```

```python
d_ord_ds_prod
```

## Display the names of all the employees in Boston.

```python
cur.execute("""SELECT firstName, lastName FROM employees 
            JOIN offices
            on employees.officeCode = offices.officeCode;""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
```

```python
df
```

```python
#What does the employee table even look like?
cur.execute("""SELECT * FROM offices
            JOIN employees
            USING(officeCode)
            
;""")
em_of = pd.DataFrame(cur.fetchall())
em_of
```

## Do any offices have no employees?
